# Regression Model

```python
x = Dataset.iloc[:,:-3].values     #Features Matrix
y = Dataset.iloc[:,-3].values      #dependent variable vector 
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.2, random_state= 1)
```

```python
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor(n_estimators = 100, random_state = 0)
regressor.fit(x_train, y_train)
```

```python
y_pred = regressor.predict(x_test)
```

```python
np.set_printoptions(precision = 2)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

```python
from sklearn.metrics import r2_score
print('RÂ²_Score: {:.2f}%'.format(r2_score(y_test, y_pred)*100))
```

```python
from sklearn.model_selection import cross_val_score                         # Applying  K-Fold Cross Validation
accuracies = cross_val_score(estimator = regressor, X = x_train,y = y_train, cv=2)  
```

```python
print("Accuracy: {:.2f} %".format(accuracies.mean()*100))
print("Standard Deviation: {:.2f} %".format(accuracies.std()*100))
```

```python
plt.rcParams['figure.figsize'] = (15,5)
plt.plot(y_test,label = 'Actual_consume', color = 'b')
plt.plot(y_pred,label = 'Rendom_forest_regression_Predictions', color = 'g')
plt.legend(loc = "upper left")
plt.show()
```
